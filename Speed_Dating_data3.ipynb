```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.display.max_columns=200 
```

```python
dataset= pd.read_csv("Speed_Dating_Data.csv", encoding='ISO-8859-1')
```

```python
dataset
```

```python
fig_dims = (10,10)
fig = plt.subplots(figsize = fig_dims)
sns.barplot(x='age', y='date', data=dataset)
```

```python
fig_dims = (10,10)
fig = plt.subplots(figsize = fig_dims)
sns.barplot(x='gender', y='date', data=dataset)
```

```python
dataset1=dataset[["gender","wave","round","match","age_o","attr","prob","race_o","pf_o_att","date","attr_o","imprelig","race","dec_o","from","goal","samerace","career","field"]]
dataset1
```

```python
dataset1 = dataset1.rename(columns={"gender": "Gender", 
                    "age_o":"age of partner",
                     "race_o":"race of partner",
                     "pf_o_att":"partner stated preference",
                     "dec_o":"partner decision",
                     "attr_o":"partner rating of attributes",
                     "imprace":"race importance",
                     "imprelig":"religion importance",
                     "from":"originally from",
                     "date":"frequency of date"})

dataset1
```

```python
dataset1["career"]=dataset1["career"].apply(lambda x: 
                                           "lawyer"              if x==1
                                       else"Academic / Research" if x==2
                                       else "Psychologist"       if x==3
                                       else"Doctor/Medicine"     if x==4
                                       else "Engineer"           if x==5
                                       else" Creative Arts/ Entertainment" if x==6 
                                       else "Consulting /Finance / Marketing/ Business " if x==7
                                       else "Real Estate"       if x==8 
                                       else"International /Humanitarian Affairs" if x==9
                                       else"Undecided"          if x ==10
                                       else"Social Work"        if x==11
                                       else "Speech Pathology"  if x==12
                                       else "Politics"          if x==13
                                       else "Journalism"        if x==14
                                       else  "Architecture"     if x==15
                                       else  "Other"            if x==16
                                           else x)                        

                
dataset1["match"]=dataset1["match"].apply(lambda x:
                                        "yes" if x==1
                                    else"no" if x==0 
                                    else x)

dataset1["goal"]=dataset1["goal"].apply(lambda x:
                               "Seemed like a fan night out" if x==1
                           else"To meet new people"          if x==2                
                           else "To get a date"              if x==3 
                           else "Looking for serious relationship" if x==4
                           else "To say I did it"            if x==5
                           else "Other"                      if x==6
                           else x )

dataset1["frequency of date"]=dataset1["frequency of date"].apply(lambda x:             
                                      "Several times a week" if x==1
                                 else"Twice a week"          if x==2
                                 else"Once a week"           if x==3
                                 else"Twice a month"         if x==4
                                 else"Once a month"          if x==5
                                 else"Several times a year"  if x==6 
                                 else "Almost never"         if x==7
                                 else x)

dataset1["race"]=dataset1["race"].apply(lambda x:
                                    "Black/African American"      if x==1
                                else"European/Caucasion-American" if x==2
                               else"Latino/Hispanic American"     if x==3
                               else"Asian/Pacific Islander/Asian-American" if x==4
                               else"Native American"              if x==5
                               else "Other"                       if x==6 
                               else x)

dataset1["Gender"]=dataset1["Gender"].apply(lambda x:
                                         "femele"   if x==1
                                     else"male"    if x==0
                                     else x)

dataset1["partner decision"]=dataset1["partner decision"].apply(lambda x:
                                                  "yes"   if x==1
                                              else"no"    if x==0  
                                               else x)                     
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Gender',y='wave', data=dataset1)
```

```python
sns.countplot(dataset1['race'])
```

```python
sns.set(style="whitegrid", color_codes=True)
sns.countplot(x="Gender", hue= "match", data=dataset1)
plt.xlabel('Gender')
plt.title('Distribution of Matches by Gender')
```

```python
plt.subplots(figsize=(6,8))
sns.set(style="whitegrid", color_codes=True)
sns.countplot(x="Gender", hue= "partner decision", data=dataset1)
plt.xlabel('Gender')
plt.title('Acceptance by Date for Gender')
```

```python
(
dataset1
    .groupby(["race"])
    [["match", "Gender"]]
   .count()
    .sort_values("match", ascending=False)
    .head(100)
).plot.pie(y="Gender", autopct='%1.0f%%')
```

```python
prob_attr = sns.lmplot(x = 'prob', y = 'attr', data = dataset1, height = 6, aspect = 1.5)
plt.xlabel('Attractiveness')
plt.ylabel('Probability of a Second Date')
plt.title('Probability of a Second Date vs. Attractiveness', fontsize = 16)
plt.grid()
plt.show()
```

```python
plt.subplots(figsize=(20,15))
ax = plt.axes()
ax.set_title("Correlation Heatmap")
corr = dataset1.corr()
corrplot=sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
corrplot.set_xticklabels(corrplot.get_xticklabels(), rotation = 45, horizontalalignment = 'right') #For X axis labels
plt.show()
```
